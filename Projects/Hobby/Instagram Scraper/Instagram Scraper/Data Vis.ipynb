```python
import pandas as pd
```

```python
import matplotlib
```

```python
from matplotlib import pyplot
```

```python
from datetime import datetime, date
```

```python
data = pd.read_csv("subs_date.csv")
```

```python
selenagomez = data[data.profile == "selenagomez"]
```

```python
kyliejenner = data[data.profile == "kyliejenner"]
```

```python
therock = data[data.profile == "therock"]
```

```python
instagram = data[data.profile == "instagram"]
```

```python
drake = data[data.profile == "champagnepapi"]
```

```python
jcole = data[data.profile == "realcoleworld"]
```

```python
dates = matplotlib.dates.date2num(selenagomez.date)
x = range(len(selenagomez.subs))
pyplot.plot(dates, selenagomez.subs, "o")
pyplot.plot(dates, kyliejenner.subs, "o")
pyplot.plot(dates, therock.subs, "o")
pyplot.plot(dates, instagram.subs, "o")
pyplot.xlabel("date and time")
pyplot.ylabel("subs")
pyplot.title("subs total")
pyplot.legend(["selenagomez", "kyliejenner", "therock", "instagram"])
pyplot.show()
```

```python
dates = matplotlib.dates.date2num(selenagomez.date)
pyplot.plot(dates, selenagomez.subs/selenagomez.subs.iloc[0]*100)
pyplot.plot(dates, kyliejenner.subs/kyliejenner.subs.iloc[0]*100)
pyplot.plot(dates, therock.subs/therock.subs.iloc[0]*100)
pyplot.plot(dates, drake.subs/drake.subs.iloc[0]*100)
pyplot.plot(dates, instagram.subs/instagram.subs.iloc[0]*100)
pyplot.plot(dates, jcole.subs/jcole.subs.iloc[0]*100)
pyplot.xlabel("date and time")
pyplot.ylabel("sub gain")
pyplot.title("subs gain in %")
pyplot.legend(["selenagomez", "kyliejenner", "therock", "drake", "instagram", "jcole"])
pyplot.show()
```

```python
sg_liste = []
for x in range(len(dates)-1):
    sg_liste.append(selenagomez.subs.iloc[x+1]/selenagomez.subs.iloc[x]*100)
kj_liste = []
for x in range(len(dates)-1):
    kj_liste.append(kyliejenner.subs.iloc[x+1]/kyliejenner.subs.iloc[x]*100)
insta_liste = []
for x in range(len(dates)-1):
    insta_liste.append(instagram.subs.iloc[x+1]/instagram.subs.iloc[x]*100)
tr_liste = []
for x in range(len(dates)-1):
    tr_liste.append(therock.subs.iloc[x+1]/therock.subs.iloc[x]*100)
drake_liste = []
for x in range(len(dates)-1):
    drake_liste.append(drake.subs.iloc[x+1]/drake.subs.iloc[x]*100)
jc_liste = []
for x in range(len(dates)-1):
    jc_liste.append(jcole.subs.iloc[x+1]/jcole.subs.iloc[x]*100)

pyplot.plot(dates[1:], sg_liste, "o")
pyplot.plot(dates[1:], kj_liste, "o")
pyplot.plot(dates[1:], tr_liste, "o")
pyplot.plot(dates[1:], drake_liste, "o")
pyplot.plot(dates[1:], insta_liste, "o")
pyplot.plot(dates[1:], jc_liste, "o")
pyplot.legend(["selenagomez", "kyliejenner", "therock", "drake", "instagram", "jcole"])
pyplot.show()
```
